```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/Hospital_Revised_Flatfiles/Payment.csv')
```

```python
df.columns
```

```python
df = df.drop(['Provider ID', 'Address', 'City', 'ZIP Code', 'County name', 'Payment category', 'Phone number', 'Value of care footnote', 'Value of care display ID', 'Value of care category','Lower estimate', 'Higher estimate', 'Payment footnote', 'Measure start date', 'Measure end date'], axis = 1)
```

```python
for row in df.itertuples():
    if row[6] != 'Not Available':
        df.loc[row.Index, 'Payment'] = float(row[6].replace("$","").replace(",",""))
    else:
        if row[4] == 'PAYM_30_AMI':
            df.loc[row.Index, 'Payment'] = 23119
        elif row[4] == 'PAYM_30_HF':
            df.loc[row.Index, 'Payment'] = 16190
        elif row[4] == 'PAYM_90_HIP_KNEE':
            df.loc[row.Index, 'Payment'] = 17026
        elif row[4] == 'PAYM_30_PN':
            df.loc[row.Index, 'Payment'] = 22567
```

```python
list1 = [1]*19224
```

```python
df['Category'] = list1
```

```python
df.Payment = df.Payment.astype(float)
```

```python
gf = df.pivot_table(values='Payment', index=['State', 'Hospital name'], columns=['Category'], aggfunc = np.mean)
```
